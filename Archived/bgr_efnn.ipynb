```
% tensorflow_version 2.x
import numpy as np
import tensorflow as tf
```

```
# data = np.load('./rgb_data.npy')
# labels = np.load('./rgb_labels.npy')

data = np.load('./rgb_data_3rd_2evs.npy')
labels = np.load('./rgb_labels_3rd_2evs.npy')

print(data[0], labels[0])
```

```
idx = np.random.permutation(len(data))
data = data[idx]
labels = labels[idx]

print(data[0], labels[0])
```

```
data = data / 255
labels = labels / 255

print(data[0], labels[0])
```

```
limit = (len(data) // 5) * 4

train_data, test_data = data[:limit], data[limit:]
del data

train_labels, test_labels = labels[:limit], labels[limit:]
del labels
```

```
model = tf.keras.Sequential()

model.add(tf.keras.layers.Dense(5, input_shape = (6,), activation='linear'))
model.add(tf.keras.layers.Dense(4, activation='linear'))
model.add(tf.keras.layers.Dense(3, activation='sigmoid'))

# model = tf.keras.models.load_model("bgr_model.hdf5")

model.compile(loss='mean_squared_error', optimizer= tf.keras.optimizers.Adam(lr=0.0001))
```

```
checkpoint = tf.keras.callbacks.ModelCheckpoint("2hidden_model_2evs.hdf5", monitor='loss', verbose=1, save_best_only=True, mode='auto', save_freq = 'epoch')

model.fit(train_data, train_labels, epochs = 100, batch_size = 2048, callbacks=[checkpoint])
```

```
res = model.evaluate(test_data, test_labels)
print(res)
# print(train_data.shape)
```
