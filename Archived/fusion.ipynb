```
import os
SAMPLES_DIR = "/content/gdrive/My Drive/efnn/dataset/jpg/"

samples = os.listdir(SAMPLES_DIR)
samples.sort()
print(samples)
```

```
# PRESETS
ALL_POSSIBLE_EVS = ['-8', '-7', '-6', '-5', '-4', '-3', '-2', '-1', '+0', '+1', '+2', '+3',]
EF_EVS = ['-8', '-4', '+1',]
EF1_EVS = ['-8', '-7', '-6', '-5', '-4', '-3', '-2', '-1', '+0',]
EF2_EVS = ['-6', '-2', '+1',]
EF3_EVS = ['-7', '-3', '+1',]
EF4_EVS = ALL_POSSIBLE_EVS
EF5_EVS = ['-8', '-5', '-2',]
```

```
from tqdm.notebook import tqdm
from efnn.efimg import Exposures

print("CREATING EF")
for sample in tqdm(samples):
    Exposures(
        os.path.join(SAMPLES_DIR, sample), 
        evs = [f'{sample}-EV{ev}' for ev in EF_EVS], 
        colorspace="bgr", 
        ef_file_name = 'ef.jpg'
    ).create_ef(check_exist = True)
```
